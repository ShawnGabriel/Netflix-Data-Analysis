### Most Popular Release Year for Movies

```python
movie_df = df[df['type'] == 'Movie']
```

```python
movie_df['count'] = 1
release_years_summed = movie_df.groupby(['release_year']).sum().reset_index()[['release_year', 'count']]
```

```python
release_years_summed.sort_values(['count'], ascending=False)
```

### Year When Netflix Added the Most Content

```python
df.info()
```

```python
df['new_date'] = pd.to_datetime(df['date_added'], format = '%B %d, %Y', errors = 'coerce')

df['date_added_year'] = df['new_date'].dt.year

df.head(2)
```

```python
df['count'] = 1
date_added_years_summed = df.groupby(['date_added_year'])['count'].sum().reset_index()[['date_added_year', 'count']]
date_added_years_summed.sort_values(['count'], ascending=False)
```

### Most Popular Month to Add New Content, Independent of Year

```python
df['date_added_month'] = df['new_date'].dt.month
df.head(2)
```

```python
date_added_months_summed = df.groupby(['date_added_month'])['count'].sum().reset_index()[['date_added_month', 'count']]
date_added_months_summed.sort_values(['count'], ascending=False)
```

### Movie With Longest and Shortest Title in The Dataset

```python
df.head()
movie_df = df[df['type'] == 'Movie']
```

```python
movie_df['title_length'] = movie_df['title'].apply(lambda x: len(x))
```

```python
top_title_lengths = movie_df.sort_values(['title_length'], ascending = False)
print("Longest title:", top_title_lengths.iloc[0]['title'])
print("Shortest title:", top_title_lengths.iloc[-1]['title'])
```

### The Actor/Actress with Most Frequent Appearance

```python
title_and_cast_df = df[['title', 'cast']].dropna()
title_and_cast_df.head()
```

```python
cast_names_counter = {}

for index, row in title_and_cast_df.iterrows():
  movie_cast = row['cast']
  movie_cast_split = movie_cast.split(', ')
  for name in movie_cast_split:
    if name in cast_names_counter:
      cast_names_counter[name] += 1
    else:
      cast_names_counter[name] = 1
```

```python
dict(sorted(cast_names_counter.items(), key=lambda item: item[1], reverse=True))
```
